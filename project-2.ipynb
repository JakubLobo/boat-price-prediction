```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.pandas.set_option('display.max_columns',None)
df=pd.read_csv("Boats_Cleaned_dataset.csv")
df.head()
```

```python
df[['year','maxEngineYear','minEngineYear','created_date','created_year']]
```

```python
# list of numerical variables
numerical_features = [feature for feature in df.columns if df[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))

# visualise the numerical variables
df[numerical_features].head()
```

```python
# list of variables that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
for feature in ['maxEngineYear','minEngineYear','year']:
  df[feature]=df['created_year']-df[feature]
df.head(50)
```

```python
df[['maxEngineYear','minEngineYear','year']].head()
```

```python
import numpy as np
num_features=['year','length_ft', 'beam_ft', 'dryWeight_lb','numEngines','totalHP','maxEngineYear','minEngineYear','sellerId','price','created_month','created_year']

for feature in num_features:
    
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

```python
discrete_feature = []

for feature in numerical_features:
    if len(df[feature].unique()) < 25 and feature not in year_feature + ['id']:
        discrete_feature.append(feature)

print("Discrete Variables Count:", len(discrete_feature))
```

```python
categorical_features = []

for feature in df.columns:
    if df[feature].dtype == 'O':
        categorical_features.append(feature)

print("Categorical Variables Count:", len(categorical_features))
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=df.groupby(feature)['price'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
```

```python
df.head(10)
```

```python
#Label Encoding
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['price'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
#scaling_feature=[feature for feature in df.columns if feature not in ['id','price'] ]
#len(scaling_feature)
scaling_feature = []

for feature in df.columns:
    if feature not in ['id', 'price']:
        scaling_feature.append(feature)

print("Number of Scaling Features:", len(scaling_feature))
```

```python
feature_scale = []

for feature in df.columns:
    if feature not in ['id', 'price']:
        feature_scale.append(feature)

print("Number of Features to Scale:", len(feature_scale))


from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[feature_scale])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
```

```python
X = df.drop(["id","price"], axis=1)
y = df["price"]
```

```python
# Assuming x and y are your feature and target variables
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Create a Random Forest Regressor
rf_regressor = RandomForestRegressor()

# Define the grid of hyperparameters to search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
#     'min_samples_split': [2, 5, 10],
#     'min_samples_leaf': [1, 2, 4]
}

# Perform Grid Search to find the best hyperparameters
grid_search = GridSearchCV(estimator=rf_regressor, param_grid=param_grid, cv=10, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
y_pred = grid_search.predict(X_test)

mean_absolute_error(y_test, y_pred)
```
